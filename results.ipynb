```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import sys
sys.path.append("../Betti-matching-3D/build")
import betti_matching as bm
sys.path.append("build")
import morse_complex as mc
```

```python
path_data = "topcow_ct"
```

```python
df_f = pd.read_json(path_data + "/foregrounds/result.json", orient="index")
df_b = pd.read_json(path_data + "/backgrounds/result.json", orient="index")
```

### Number of examples with critical n-cells below 1:

```python
print(len(df_f[df_f['< 1.000000 in dim 0'] > 0]))
print(len(df_f[df_f['< 1.000000 in dim 1'] > 0]))
print(len(df_f[df_f['< 1.000000 in dim 2'] > 0]))
print(len(df_f[df_f['< 1.000000 in dim 3'] > 0]))
```

### Examples whose number of critical cells is not minimal wrt to persistent betti numbers:

```python
examples = []
for i in range(len(df_f)):
    example = df_f.iloc[i]
    image = np.load(path_data + "/foregrounds/" + example.name)

    BM = bm.BettiMatching(image, image)
    BM.compute_matching()
    m = BM.get_matching()

    if (len(m[0][0])+1 != example["< 1.000000 in dim 0"] or 
        len(m[0][1]) != example["< 1.000000 in dim 1"] or 
        len(m[0][2]) != example["< 1.000000 in dim 2"]):
        examples.append(example.name)
        print(example.name)
```
